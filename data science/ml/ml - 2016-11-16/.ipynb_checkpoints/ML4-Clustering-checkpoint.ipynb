Prepare a sinthetic dataset for this example.

```python
from sklearn import datasets

X, _ = datasets.make_blobs(n_samples=150, random_state=8, centers=4)
# X, _ = datasets.make_circles(n_samples=150, factor=.5, noise=.05)

plot(X[:, 0], X[:, 1], 'bx')
#cercheremo l'algoritmo che clusterizza queste 4 nuvole di punti
```

# Kmeans clustering

```python
np.random.seed(42)

from sklearn import cluster

#dobbiamo indicare il numero di cluster in partenza
#vediamo come si comporta con il nostro data set
#random --> per iniziare l'algoritmo prendo un numero n di punti a caso
#quindi cambiando il seed la clusterizzazione varia, in genere con kmeans si usa quindi una media di un gran
#numero di tentativi
kmeans = cluster.KMeans(n_clusters=4, init='random',
                        n_init=1, max_iter=10)
#fa il fit sui dati e ritorna un predict con il cluster per ciascun punto
C = kmeans.fit_predict(X)

colors = array(['b', 'r', 'g', 'k', 'y'])
scatter(X[:, 0], X[:, 1], marker="x", c=colors[C])

centers = kmeans.cluster_centers_
scatter(centers[:, 0], centers[:, 1], s=100, c=colors)
#posso valutare la bonta' della clusterizzazione con la distanza media fra i punti di ciascun cluster e
#il relativo centro
```

## Hierarchical clustering

```python
X, _ = datasets.make_blobs(n_samples=50, random_state=8, centers=4)
# X, _ = datasets.make_circles(n_samples=200, factor=.1, noise=.05)

scatter(X[:, 0], X[:, 1]);
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(X, method='average', metric='euclidean')
figure(figsize=(15,5))
P = dendrogram(Z); #e' l'albero creato dalla gerarchia dell'albero costruito bottom-up
#a seconda del livello cui "taglio" l'albero ottengo 1-2-3-4-... cluster mano a mano che abbasso il livello
#in questo caso sto usando il metodo 'average', ovvero considero la distanza media
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(X, method='complete', metric='euclidean')
figure(figsize=(15,5))
P = dendrogram(Z); 
```

```python
#prende l'albero prima creato e restituisce 4 cluster
T = fcluster(Z, 4, 'maxclust')

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[T-1]);
```

### Hierarchical Clustering in Scikit-Learn:

```python
from sklearn import cluster

model = cluster.AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')

clusters = model.fit_predict(X)

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[clusters-1]);
```

Test the different clustering approaches with a "circles" dataset.

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(X, method='single', metric='euclidean')
#figure(figsize=(15,5))
#P = dendrogram(Z);
T = fcluster(Z, 2, 'maxclust')

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[T-1]);
```
