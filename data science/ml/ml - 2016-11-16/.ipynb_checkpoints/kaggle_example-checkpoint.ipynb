# Feature Extraction

```python
subreddits = train_data.subreddit.unique()
subreddits_map = pd.Series(index=subreddits, data=arange(subreddits.shape[0]))
```

```python
subreddits.shape
```

```python
subreddits_map
#individuo la mappatura dei subreddit che mi servira' come indicizzazione per la matrice sparsa
```

```python
#carico il pacchetto per creare le matrici sparse
from scipy import sparse
```

```python
#definisco la funzione che estrae le features di un determinato gruppo
#la funzione genera il gruppo dei commenti dello stesso autore
def extract_features(group):
    group_subreddits = group['subreddit'].values
    idxs = subreddits_map[group_subreddits].values
    #utilizzo una dictionary of keys sparse matrix
    #ci consente di creare la matrice senza inizializzare tutte le entrate della matrice
    #devo creare un dizionario per ciascuna entrata non nulla della matrice
    #non  la matrice sparsa piu' ottimizzata, ma e' quella utile per fare tutte le trasformazioni matriciali
    v = sparse.dok_matrix((1, subreddits.shape[0]))
    #definisco le dimensioni
    for idx in idxs:
        if not np.isnan(idx):
            v[0, idx] = 1
    return v.tocsr()

extract_features(train_data[train_data.author=='savoytruffle'])
```

```python
features_dict = {}
#creo un dizionario per ciascun autore
#invece di selezionare tutti i commenti di quell'utente, li raggruppo sia per subreddit che per autore
for author, group in train_data.groupby('author'):
    features_dict[author] = extract_features(group)
```

```python
#per ogni autore prendo la feature vector e creo la matrice
#con vstack trasformo ogni matrice in una commpressed sparse raw matrix
X = sparse.vstack([features_dict[author] for author in target.author])
X
```

```python
y = target.gender
```

# Model Selection

```python
from sklearn import metrics
#provare con altri modelli gia' visti
```

```python
from sklearn import linear_model, grid_search

model = linear_model.LogisticRegression(penalty='l1')

Cs = np.logspace(-2, 0, 40)
#grid search ottimizza lo score di default che e' quello dell'accuracy, 
#invece noi dobbiamo massimizzare lo score di AuC
#gs = grid_search.GridSearchCV(model, {'C': Cs}, cv=10)
gs = grid_search.GridSearchCV(model, {'C': Cs}, cv=10, scoring = 'roc_auc')
gs.fit(X, y)
```

# Prepare the solution

```python
test_data = pd.read_csv("test_data.csv.gz", compression="gzip", encoding="utf8")
```

```python
features_dict = {}

for author, group in test_data.groupby('author'):
    features_dict[author] = extract_features(group)
```

```python
X_test = sparse.vstack([features_dict[author] for author in test_data.author.unique()])
X_test
```

```python
y_pred = model.predict_proba(X_test)[:,1]
```

```python
solution = pd.DataFrame({"author":test_data.author.unique(), "gender":y_pred})
solution.head()
```

```python
solution.to_csv("logistic_regression.csv", index=False)
```
